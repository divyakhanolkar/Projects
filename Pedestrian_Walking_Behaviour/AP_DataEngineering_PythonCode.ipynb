```python
import math


def rotate_vector(curr_x,curr_y,vector,speed,angle_degrees, velocity):
    angle_radians = math.radians(angle_degrees)
    
    x = vector[0]
    y = vector[1]
    
    if math.hypot(x,y)!= 0:
        rotated_x = (x * speed * math.cos(angle_radians) - y * speed * math.sin(angle_radians))/(math.hypot(x,y))
        rotated_y = (x * speed * math.sin(angle_radians) + y * speed * math.cos(angle_radians))/(math.hypot(x,y))
    if math.hypot(x,y)== 0:
        rotated_x = 0
        rotated_y = 0
       
    return (curr_x+(rotated_x*velocity), curr_y+(rotated_y*velocity))
```

```python
def dot_mul(curr_x,curr_z,dest_x,dest_z,alternative_x,alternative_z):
    a1=alternative_x-curr_x
    a2=dest_x-curr_x
    b1=alternative_z-curr_z
    b2=dest_z-curr_z
    if ((math.hypot(a1,b1))*(math.hypot(a2,b2))!=0):
        cos_of_Angle= (a1*a2+b2*b1)/((math.hypot(a1,b1))*(math.hypot(a2,b2)))
    elif ((math.hypot(a1,b1))*(math.hypot(a2,b2))==0):
        cos_of_Angle= 1
    Angle= (math.acos(cos_of_Angle))
    return Angle
```
